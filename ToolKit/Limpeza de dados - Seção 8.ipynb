### Missing values

```python
datas = pd.date_range('20200101', periods = 60, freq='D')
df = pd.DataFrame(np.random.randn(60,5), index=datas, columns=list('ABCDE'))
df.head()
```

```python
df['F'] = df.A[df.A>0]
df.head(10)
```

```python
df2 = df.copy()
df3 = df.copy()
```

```python
df2.dropna() # remove a linha inteira com algum valor NA
```

```python
df2.dropna().shape
```

```python
df3.head()
```

```python
df3.F.fillna(np.mean(df3.A)).head()
```

```python
df4=df.copy()
df4.F.fillna(77777)
```
