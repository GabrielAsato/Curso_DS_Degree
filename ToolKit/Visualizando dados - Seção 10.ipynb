### Estilo de plotagens

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Inverter as barras e colocar a extensão
titanic = 'C:/Users/gabri/OneDrive/Documentos/Python Scripts/titanic dataset/train.csv'
```

```python
df_titanic = pd.read_csv(titanic)
```

```python
df_titanic.head()
```

#### Settings the plot

```python
plt.rcParams['figure.figsize']=(12,4)
```

```python
df_titanic.shape
```

#### Fazendo alterações no plot

- Utilize "tab" e "Shift+tab" para ver as funções e os argumentos, respectivamente.

```python
plt.plot(df_titanic.Age,'bo')
plt.title('Informações dos passageiros')
plt.xlabel('Passageiro',size=14)
plt.ylabel('Idade',size=14)
plt.show()
```

```python
df_titanic.Age.plot(xlabel='Passageiros',ylabel='Idade')
```

```python
df_titanic.plot()
plt.show()
```

### Scatter plot

```python
plt.scatter(df_titanic.PassengerId, df_titanic.Age,c='red',s=4,marker='+')
plt.show()
```

#### Histogramas
