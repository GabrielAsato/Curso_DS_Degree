## Função pivot()

```python
df = pd.DataFrame({'foo': ['one', 'one', 'one', 'two', 'two',
                           'two'],
                   'bar': ['A', 'B', 'C', 'A', 'B', 'C'],
                   'baz': [1, 2, 3, 4, 5, 6],
                   'zoo': ['x', 'y', 'z', 'q', 'w', 't']})
df.groupby(['bar','foo']).sum().T
```

```python
pd.pivot?
#pd.pivot(DataFrame,index=,columns=,values=)
```

Um pd.DataFrame ou uma pd.Series é composto de index (rows), columns e values.

```python
pd.pivot(df,'foo','bar','baz')
```

```python
# Exemplo da aula
dias = pd.date_range('20190101',periods=12)
dias
```

```python
pessoas = ['George','Victor','Lucas']
```

```python
np.random.choice(pessoas)
```

```python
nome = []
gasto = []
for i in range(12):
    nome.append(np.random.choice(pessoas))
    gasto.append(np.round(np.random.rand()*100,2))
nome
```

```python
gasto
```

```python
df = pd.DataFrame({'Dia':dias,'Nome':nome,'Gasto':gasto})
df
```

```python
df[df.Dia == '2019-01-05']
```

```python
df.pivot('Dia','Nome','Gasto').replace(np.nan,'--')
```

```python
pd.DataFrame(df.groupby('Nome')['Gasto'].sum())
```

## Função pivot_table()

```python
Carros = [7,4,3,2,8]
dias = pd.date_range('20190101','20190101',periods=5)
vendedor = ['George','Vagner','Pedro','Vagner','George']

df = pd.DataFrame({'Vendas':Carros,
                   'Data':dias,
                   'Vendedor':vendedor})
df
```

```python
pd.pivot_table(df,index='Data',columns='Vendedor',values='Vendas',aggfunc='sum')
```

## Função melt()

```python
df = pd.DataFrame({'A':{0:'a',1:'b',2:'c'},
                   'B':{0:1, 1:3, 2:5},
                   'C':{0:2, 1:4,2:6}
                  })
```

```python
df
```

```python
pd.melt(df,id_vars = 'A', value_vars = 'B')
```

```python
pd.melt(df,id_vars = 'A', value_vars = ['B','C'],value_name='points',var_name = 'Categoria')
```

```python
data = {'localização':['Cidade_A','Cidade_B'],
        'temperatura': ['Prevista','Atual'],
        'Set-2019':[30,32],
        'Out-2019':[45,43],
        'Nov-2019':[24,22]}
data
```

```python
df = pd.DataFrame(data,columns=['localização','temperatura','Set-2019','Out-2019','Nov-2019'])
df
```

```python
# Obtenção de uma série temporal
df2 = pd.melt(df,id_vars=['localização','temperatura'],var_name='Date',value_name = 'Value')
df2
```
