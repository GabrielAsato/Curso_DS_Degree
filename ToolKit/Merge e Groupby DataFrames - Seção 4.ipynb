```python
import pandas as pd
```

```python
# Cadastro da loja A
cadastro_a = {'Id':['AA2930','BB4563','CC2139','DE2521','GT3462','HH1158'],
              'Nome':['Victor','Amanda','Bruna','Carlos','Ricardo','Maria'],
              'Idade':[20,35,40,54,30,27],
              'CEP':['00092-0290','11111-111','22222-888','00000-999','88888-111','77777-666']
             }
cadastro_a = pd.DataFrame(cadastro_a, columns = ['Id','Nome','Idade','CEP'])
cadastro_a
```

```python
# Cadstro da loja B
cadastro_b = {'Id':['CC9999','EF4488','DD9999','GT3462','HH1158'],
              'Nome':['Marcos','Patrícia','Ericka','Ricardo','Maria'],
              'Idade':[19,30,22,30,27],
              'CEP':['00092-029','11111-111','11111-888','88888-111','77777-666']
             }
cadastro_b = pd.DataFrame(cadastro_b, columns = ['Id','Nome','Idade','CEP'])
cadastro_b
```

```python
# Registro de compras de todas unidades
compras  = {'Id':['AA2930','EF4488','CC2139','EF4488','CC9999','AA2930','HH1158','HH1158'],
            'Data':['2019-01-01','2019-01-30','2019-01-30','2019-02-01','2019-02-20','2019-03-15','2019-04-01','2019-04-10'],
            'Valor':[200,100,40,150,300,25,50,500]
           }
compras = pd.DataFrame(compras, columns = ['Id','Data','Valor'])
compras
```

Iremos usar a função pd.merge()

pd.merge(tabela_na_esquerda, tabela_na_direita, on='coluna_coincidente', how='left|right|inner|outer')

```python
# Quais são todos os cadastros nas duas lojas?
pd.merge(cadastro_a,cadastro_b,how='outer')
```

```python
# Semelhante processo por pd.concat
clientes = pd.concat([cadastro_a, cadastro_b],ignore_index = True).drop_duplicates(subset='Id')
clientes
```

```python
# Quais são os cadastros em comum nos dois cadastros?
pd.merge(cadastro_a,cadastro_b,how='inner')
```

```python
esquerda = pd.merge(cadastro_a, compras, on='Id',how='left').dropna()
esquerda
```

```python
# Groupby: Agrupe o 'Id' e o 'Nome' e forneça os 'Valores' totais
esquerda_agrupada = esquerda.groupby(['Id','Nome'])['Valor'].sum()
pd.DataFrame(esquerda_agrupada)
```

```python
esq_val = esq.Valor
esq_val_list = list(esq_val)
esq_val_list
```
