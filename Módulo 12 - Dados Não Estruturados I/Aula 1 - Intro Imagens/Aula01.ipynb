## 1. O que é processamento de sinais

Processamento de sinais é uma sub-área da engenharia elétrica cujo foco é analisar, modificar, e sintetizar sinais como, som, imagens, textos, dentre outros.

Técnicas de processamento de sinais podem ser utilizadas para a melhoria de transmissão, eficiência de armazenamento, e a qualidade do sinal.

De acordo com Alan Oppenheim e Ronald Schafer, os princípios do processamento de sinal podem ser encontrados am técnicas de análise numérica clássicas do século 17. Com o desenvolvimento de computadores eletrônicos e a sua proliferação na segunda metade do século 20, houve um avanço na área de processamento de sinais. Permitindo que houvesse a aplicação dessas técnicas em um vasto número de problemas do mundo real.

## 2. Aplicações

As principais aplicações do processamento de sinais são:

- Processamento de sinal de áudio
    - Consistindo em sinais elétricos representando sons, como discursos ou música
- Processamento de imagem
- Processamento de vídeo
- Comunicação sem fio (wireless)
- Sismologia:
    - Área de estudo que busca entender a ocorrência de terremotos
- Processamento de sinais financeiros
    - Análise de dados financeiros utilizando técnicas de processamento de sinais, principalmente para propósitos de predição (forecast)
- Extração de features
- Aumento da qualidade de sinais
    - Como a redução de ruído, melhoria da imagem
- Compressão de sinal
    - Como compressão de áudio, imagem e video

```python
import cv2
mario_img = cv2.imread("mario.png")

# Não se preocupe com o código abaixo, por enquanto.
# Nele transformamos a figura de BGR para RGB
mario_img_rgb = cv2.cvtColor(mario_img, cv2.COLOR_BGR2RGB)

# Transformando a figura BGR para escala de cinza!
mario_gray_img = cv2.cvtColor(mario_img, cv2.COLOR_BGR2GRAY)

# Plotando a figura colorida, e a sua versão em escala de cinza
f, ax = plt.subplots(1, 2, figsize = (10, 5))
ax[0].imshow(mario_img_rgb)
ax[1].imshow(mario_gray_img, cmap="gray");
```

```python
gradientes={}

for bit in range(1,11):

    gradiente =  np.linspace(0, 1, 2**bit)[::-1]

    gradiente = np.vstack([gradiente for _ in range(2**bit)])

    gradientes[f"{bit}"] = gradiente.T

f, ax = plt.subplots(2, 5, figsize=(17,10))
row = 0
col = 0
for i, (key, gradiente) in enumerate(gradientes.items()):
    if i == 5: # posição dos axis dos subplots
        row = 1
        col = 0
    # Gráfico com gradiente de cores
    ax[row][col].imshow(gradiente, cmap='gray')
    
    # título do subplot
    ax[row][col].title.set_text(f'{key}-bit - {2**(int(key))} tons')

    # Remover os ticks (bom ter para ter uma visão qualitativa de pixels)
    ax[row][col].get_xaxis().set_visible(False)
    ax[row][col].get_yaxis().set_visible(False)
    col += 1

f.tight_layout()

    
```
