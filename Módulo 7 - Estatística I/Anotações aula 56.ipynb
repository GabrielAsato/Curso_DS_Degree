### O que √© rvs?
rvs (random variate) √© o resultado particular de uma distribui√ß√£o de vari√°vel aleat√≥ria. Fornece dados de tamanho espec√≠fico (size = ) que tem distribui√ß√£o normal.

```python
data = stats.norm.rvs(loc = 0, scale = 1, size = 1000)
```

```python
plt.hist(data,bins = 50)
```

```python
data = poisson.rvs(20, size= 1000)
ax = sns.displot(data,
                  bins=30,
                  kde=False,
                  color='skyblue',
                  )
ax.set(xlabel='Distribui√ß√£o Poisson', ylabel='Frequ√™ncia');
```

```python
r = stats.norm.rvs(size = 1000)
plt.plot(r)
```
