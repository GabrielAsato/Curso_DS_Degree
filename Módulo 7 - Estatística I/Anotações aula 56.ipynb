### O que é rvs?
rvs (random variate) é o resultado particular de uma distribuição de variável aleatória. Fornece dados de tamanho específico (size = ) que tem distribuição normal.

```python
data = stats.norm.rvs(loc = 0, scale = 1, size = 1000)
```

```python
plt.hist(data,bins = 50)
```

```python
data = poisson.rvs(20, size= 1000)
ax = sns.displot(data,
                  bins=30,
                  kde=False,
                  color='skyblue',
                  )
ax.set(xlabel='Distribuição Poisson', ylabel='Frequência');
```

```python
r = stats.norm.rvs(size = 1000)
plt.plot(r)
```
